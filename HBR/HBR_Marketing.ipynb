# Avoid the Pitfalls of A/B Testing   

1. <b>Not looking beyond the average  
    
    1) Use metrics and approchaes that reflect the value of different customer segments.   
    2) Measure the impact across different levels of digital access.   
    3) Always account for group-specific behaviour   
    4) Segment key markets.
    
2. <b>Forgetting that customers are connected   
    
    1) Use network A/B testing   
    2) Use time-series experiments   
    
3. <b> Focusing predominantly on the short term
    
    1) Get the legnthe of experiments right.   
    2) Run "holdout"experiments

```python
import pandas as pd
import numpy as np
```

```python
prod = pd.read_csv('project/01.Product.csv')
master = pd.read_csv('project/06.Master.csv')
```

```python
prod['PD_BRA_NM'] = prod.PD_BRA_NM.str.replace('[', '').str.replace(']', '').str.replace('.', '').str.replace(',', '').str.replace('\(.*\)', '').str.replace(' ', '')
prod.info()
```

```python
prod_mod = prod

# 브랜드에서 숫자 제거
tmp = pd.to_numeric(prod_mod['PD_BRA_NM'], errors='coerce')
tmp = tmp.dropna()
prod_mod = prod_mod.drop(index = tmp.index)

# 구매수, 가격을 숫자로
prod_mod['PD_BUY_AM'] = prod_mod['PD_BUY_AM'].str.replace(',', '')
prod_mod['PD_BUY_AM'] = pd.to_numeric(prod_mod['PD_BUY_AM'])

tmp = prod_mod['PD_BUY_CT'].str.replace(',', '')
tmp = tmp.dropna()
prod_mod['PD_BUY_CT'][tmp.index] = tmp
prod_mod['PD_BUY_CT'] = pd.to_numeric(prod_mod['PD_BUY_CT'])
```

```python
del prod_mod['PD_ADD_NM']

prod_mod
```

```python
prod_mod.to_csv('project/new_product.csv', sep=',', encoding='utf-8')
```

```python
#마스터 파일에 브랜드, 가격, 구매수
new_master = pd.merge(master, prod_mod[['PD_C', 'PD_BRA_NM', 'PD_BUY_AM']].drop_duplicates('PD_C'), on='PD_C', how='left')
new_master = pd.merge(new_master, prod_mod.groupby('PD_C')[['PD_BUY_CT']].sum(), on='PD_C', how='left')

#인트값(안해도됨)
new_master['PD_BUY_CT'] = new_master['PD_BUY_CT'].fillna(0)
new_master['PD_BUY_CT'] = new_master['PD_BUY_CT'].astype('int')
new_master['PD_BUY_AM'] = new_master['PD_BUY_AM'].fillna(0)
new_master['PD_BUY_AM'] = new_master['PD_BUY_AM'].astype('int')
new_master
```

```python
new_master.to_csv('project/new_master.csv', sep=',', encoding='utf-8')
```

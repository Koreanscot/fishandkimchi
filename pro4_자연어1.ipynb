```python
import pandas as pd
import numpy as np
import re
```

```python
master = pd.read_csv('project/new_master.csv', index_col=0)
master.head()
```

```python
master['strings'] = master['PD_NM'] + ' ' + master['PD_BRA_NM'].fillna('') + ' ' + master['CLAC3_NM']
```

```python
master['strings'] = [re.sub(r'\W|\d|[A-Z]|[a-z]|_', ' ', master['strings'][n]) for n in range(len(master))]
master.head()
```

```python
prod = pd.read_csv('project/new_product.csv', index_col=0)
prod.head()
```

```python
string_prod = pd.merge(prod[['CLNT_ID', 'PD_C', 'PD_BUY_CT']], master[['PD_C', 'PD_NM', 'CLAC1_NM', 'CLAC2_NM', 'CLAC3_NM', 'strings']], on='PD_C').sort_values('CLNT_ID')
```

```python
string_prod.to_csv('project/string_prod.csv', sep=',', encoding='utf-8')
```

```python
string_prod = string_prod.set_index('CLNT_ID')
string_prod = string_prod.drop('PD_BUY_CT', axis=1).drop_duplicates()
```

```python
string_prod_pivot = string_prod.pivot_table('strings', 'CLNT_ID', aggfunc='sum')
```

```python
string_prod_pivot.to_csv('project/string_prod_pivot.csv', sep=',', encoding='utf-8')
```
